Data Preprocessing

```python
from transformers import DistilBertTokenizer

# Initialize the DistilBERT tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

# Define special tokens to prevent splitting
special_tokens_dict = {'additional_special_tokens': ['c#', 'f#', 'g#']}  # Add '#' prefixes for pitch values

# Add special tokens to the tokenizer
tokenizer.add_special_tokens(special_tokens_dict)

# Tokenize each triplet in the dataset
with open('/content/combined_dataset.txt', 'r') as file:
    dataset = file.readlines()

# Initialize lists to store tokenized triplets and their corresponding token IDs
tokenized_triplets = []
token_ids_triplets = []

# Clean and tokenize each triplet in the dataset
for line in dataset:
    # Step 2: Clean the data (remove unwanted characters)
    line = line.strip()  # Remove leading and trailing whitespaces
    line = line.replace('<', '').replace('>', '')  # Remove '<' and '>' symbols

    # Step 3: Split the line into triplets based on the delimiter (' | ')
    triplets = line.split(' | ')

    # Initialize lists to store tokenized parts of each triplet
    tokenized_parts = []
    token_ids_parts = []

    # Tokenize each triplet into pitch, duration, and rest parts
    for triplet in triplets:
        # Step 4: Tokenization
        tokens = tokenizer.tokenize(triplet)

        # Step 5: Convert tokens to IDs
        token_ids = tokenizer.convert_tokens_to_ids(tokens)

        # Append the tokenized parts and their corresponding token IDs to the lists
        tokenized_parts.append(tokens)
        token_ids_parts.append(token_ids)

    # Append the tokenized triplet and its token IDs to the respective lists
    tokenized_triplets.append(tokenized_parts)
    token_ids_triplets.append(token_ids_parts)

# Print the tokenized triplets and their corresponding token IDs
for i in range(len(tokenized_triplets)):
    print("Tokenized triplet:", tokenized_triplets[i])
    print("Token IDs:", token_ids_triplets[i])
    print()
```
